```python
import numpy as np
import pandas as pd
```

```python
banking_df = pd.read_csv("subscription_prediction.csv")
banking_df["y"] = banking_df["y"].apply(lambda x:1 if x=="yes" else 0)
```

```python
banking_df.columns
```

```python
banking_df_c=pd.get_dummies(banking_df,columns=['job','marital','education'])
```

```python
banking_df_c.head(1)
banking_df_c.columns
```

```python
X=banking_df_c.drop(['default','housing','loan','contact','month','day_of_week','poutcome','y'],axis=1)
```

```python
y=banking_df_c["y"]
```

```python
columns_t=X.columns
columns_t
```

```python
banking_df_c.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
X=scale.fit_transform(X)
```

```python
data_df=pd.DataFrame(X,columns=columns_t)
data_df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2,random_state=417)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_train,y_train,test_size=0.2*X.shape[0]/X_train.shape[0],random_state=417)
```

```python
X_train.shape[0],X_val.shape[0],X_test.shape[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=15,metric='minkowski',p=1,weights='distance')
```

```python
knn.fit(X_train,y_train)
```

```python
y_val
```

```python
prediction=knn.predict(X_val)
```

```python
(y_val==prediction).sum()/len(y_val)*100
```

```python
knn.score(X_val,y_val)*100
```

```python
knn.predict_proba(X_val)
```
